```python
import matplotlib.pyplot as plt
%matplotlib inline
import pickle
import numpy as np
```

```python
# Load data
ratings = pickle.load(open("data/ratings.p", "rb"))
movies = pickle.load(open("data/movies.p", "rb"))
```

```python
# Plot a histogram of all ratings in the dataset
scores = ratings[:,2]
plt.hist(scores, bins = [0.5, 1.5, 2.5, 3.5, 4.5, 5.5])
plt.xlabel("Rating")
plt.ylabel("Number of Ratings")
plt.title("Movie Lens Dataset All Ratings")
plt.show()
```

```python
# Plot a histogram for the ratings of the top ten most
# popular movies, most popular as being defined as having the most
# ratings
movies_scored = ratings[:,1]
movie_num_ratings = {}
for movie_id in movies_scored:
    if movie_id in movie_num_ratings:
        movie_num_ratings[movie_id] += 1
    else:
        movie_num_ratings[movie_id] = 1
movies_sorted = sorted(movie_num_ratings.items(), key=lambda x: x[1], reverse = True)
top_ten_movies = [movie_id for movie_id,_ in movies_sorted][:10]

top_ten_ratings = []
for _,movie_id,r in ratings:
    if movie_id in top_ten_movies:
        top_ten_ratings.append(r)
        
plt.hist(top_ten_ratings, bins = [0.5, 1.5, 2.5, 3.5, 4.5, 5.5])
plt.xlabel("Rating")
plt.ylabel("Number of Ratings")
plt.title("Movie Lens Dataset All Ratings For 10 Most Popular")
plt.show()
```

```python
# Plot a histogram of the ratings of the ten best movies
# "Best" as defined as the movies with the highest average
# rating

movie_scores = {}
for _, movie_id, score in ratings:
    if movie_id in movie_scores:
        movie_scores[movie_id].append(score)
    else:
        movie_scores[movie_id] = [score]

movie_avg_score = {
    movie_id: np.sum(scores) / len(scores)
    for movie_id, scores in movie_scores.items()
}

movies_sorted = sorted(movie_avg_score.items(), key=lambda x: x[1], reverse = True)
top_ten_movies = [movie_id for movie_id,_ in movies_sorted][:10]

top_ten_ratings = []
for _,movie_id,r in ratings:
    if movie_id in top_ten_movies:
        top_ten_ratings.append(r)

plt.hist(top_ten_ratings, bins = [0.5, 1.5, 2.5, 3.5, 4.5, 5.5])
plt.xlabel("Rating")
plt.ylabel("Number of Ratings")
plt.title("Movie Lens Dataset All Ratings For 10 Best")
plt.show()
```

```python
# Ratings of movies from three genres

action_movies = []
adventure_movies = []
animation_movies = []
for movie_id,_,_, action, adventure, animation,_,_,_,_,_,_,_,_,_,_,_,_,_,_,_ in movies:
    if action == 1:
        action_movies.append(movie_id)
    if adventure == 1:
        adventure_movies.append(movie_id)
    if animation == 1:
        animation_movies.append(movie_id)
        
action_ratings = []
for _,movie_id,r in ratings:
    if movie_id in action_movies:
        action_ratings.append(r)
        
adventure_ratings = []
for _,movie_id,r in ratings:
    if movie_id in adventure_movies:
        adventure_ratings.append(r)
        
animation_ratings = []
for _,movie_id,r in ratings:
    if movie_id in animation_movies:
        animation_ratings.append(r)
        
plt.hist(action_ratings, bins = [0.5, 1.5, 2.5, 3.5, 4.5, 5.5])
plt.xlabel("Rating")
plt.ylabel("Number of Ratings")
plt.title("Movie Lens Dataset All Ratings For Action Movies")
plt.show()

plt.hist(adventure_ratings, bins = [0.5, 1.5, 2.5, 3.5, 4.5, 5.5])
plt.xlabel("Rating")
plt.ylabel("Number of Ratings")
plt.title("Movie Lens Dataset All Ratings For Adventure Movies")
plt.show()

plt.hist(animation_ratings, bins = [0.5, 1.5, 2.5, 3.5, 4.5, 5.5])
plt.xlabel("Rating")
plt.ylabel("Number of Ratings")
plt.title("Movie Lens Dataset All Ratings For Animation Movies")
plt.show()
```
